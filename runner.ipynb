```python
import cv2
import pathlib as path
import sys
path_root = path.Path.cwd() # Assuming the script is run from the root directory
```

```python

def mp4_to_avi(input_file, output_file):
    print(f"Getting ready for convering {input_file} to {output_file}")
    # Open the input video
    input_file = str(input_file)
    output_file = str(output_file)

    cap = cv2.VideoCapture(input_file)
    
    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    
    # Define codec and create VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_file, fourcc, fps, (width, height))
    
    # Process each frame
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        out.write(frame)
    
    # Release resources
    cap.release()
    out.release()

    print(f"Converted {input_file} to {output_file}")

input_file = path_root / 'input/raw/subject1/video1142643710.mp4'
output_file = path_root / 'input/raw/subject1/video1142643710.avi'

mp4_to_avi(input_file, output_file)
```
